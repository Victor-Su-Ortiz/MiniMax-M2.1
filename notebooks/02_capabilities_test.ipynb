# MiniMax-M2.1 Capabilities Test

This notebook explores the various capabilities of the MiniMax-M2.1 model:
- Reasoning and logic
- Code generation
- Creative writing
- Math and calculations
- Multi-turn conversation coherence


````python
# Setup
import sys
sys.path.insert(0, '..')

from dotenv import load_dotenv
load_dotenv('../.env')

from src.minimax_client import MiniMaxClient

client = MiniMaxClient()
print(f"Testing: {client.model}")
````

## 1. Reasoning and Logic

Test the model's ability to reason through problems step by step.


````python
# Logic puzzle
response = client.simple_chat("""
A farmer needs to cross a river with a wolf, a goat, and a cabbage. 
The boat can only carry the farmer and one item at a time.
If left alone:
- The wolf will eat the goat
- The goat will eat the cabbage

How can the farmer get everything across safely?
""")
print("Logic Puzzle Solution:")
print(response)
````

````python
# Deductive reasoning
response = client.simple_chat("""
All roses are flowers.
Some flowers fade quickly.
Can we conclude that some roses fade quickly? 
Explain your reasoning.
""")
print("Deductive Reasoning:")
print(response)
````

## 2. Code Generation

Test the model's ability to write and explain code.


````python
# Python code generation
response = client.simple_chat("""
Write a Python function that implements binary search on a sorted list.
Include docstring and type hints.
""")
print("Binary Search Implementation:")
print(response)
````

````python
# Code explanation
response = client.simple_chat("""
Explain what this code does and identify any potential issues:

```python
def mystery(n):
    if n <= 1:
        return n
    return mystery(n-1) + mystery(n-2)
```
""")
print("Code Explanation:")
print(response)
````

## 3. Creative Writing

Test the model's creative and narrative abilities.


````python
# Short story
response = client.simple_chat(
    "Write a short story (100-150 words) about a robot learning to paint.",
    temperature=0.9  # Higher temperature for creativity
)
print("Creative Story:")
print(response)
````

````python
# Poetry
response = client.simple_chat(
    "Write a haiku about artificial intelligence.",
    temperature=0.8
)
print("AI Haiku:")
print(response)
````

## 4. Math and Calculations

Test mathematical reasoning and problem-solving.


````python
# Word problem
response = client.simple_chat("""
A train leaves Station A at 9:00 AM traveling at 60 mph.
Another train leaves Station B (300 miles away) at 10:00 AM traveling toward Station A at 80 mph.
At what time will they meet? Show your work.
""")
print("Math Word Problem:")
print(response)
````

````python
# Algebra
response = client.simple_chat("""
Solve for x: 3x² - 12x + 9 = 0
Show your solution step by step.
""")
print("Algebra Problem:")
print(response)
````

## 5. Multi-turn Conversation Coherence

Test the model's ability to maintain context across multiple turns.


````python
# Multi-turn context test
conversation = [
    {"role": "system", "content": "You are a helpful assistant with perfect memory."}
]

# Define questions that require remembering context
questions = [
    "My name is Alex and I have 3 cats named Luna, Milo, and Shadow.",
    "What's my oldest cat's name? (Luna is the oldest)",
    "How many pets do I have in total?",
    "What's my name?"
]

print("Multi-turn Conversation Coherence Test:")
print("=" * 50)

for q in questions:
    conversation.append({"role": "user", "content": q})
    response = client.chat(conversation)
    answer = response.choices[0].message.content
    conversation.append({"role": "assistant", "content": answer})
    
    print(f"\nUser: {q}")
    print(f"Assistant: {answer}")
````
